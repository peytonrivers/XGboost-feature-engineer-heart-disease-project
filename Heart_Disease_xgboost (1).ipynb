```python
import sklearn
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model = XGBClassifier(max_depth=2, min_child_weight=3, random_state=42)
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
residual = y_test - predict
```

```python
predict
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
print("Accuracy Score is: ", accuracy_score(y_test, predict))
print("Classification Report is: ", classification_report(y_test, predict))
```

```python
results = X_test.copy()
results['Actual'] = y_test
results['Predicted'] = predict
results['Correct'] = (results['Actual'] == results['Predicted']).astype(int)
```
